# Clean parsed table

```python
# Separate city and county
web_parse[['City','County']] = web_parse['City|County'].str.split('|',expand=True)
web_parse.head()
```

```python
web_parse = web_parse.drop('City|County',axis=1)
web_parse.head()
```

```python
first_column = web_parse.pop('City')
web_parse.insert(0,'City',first_column)
web_parse.head()
```

```python
# Delete unnecessary characters
web_parse_after = web_parse.replace(['<td>','</td>','<sup id="cite_ref-4" class="reference">','<sup id="cite_ref-5" class="reference">','<a href="#cite_note-5">[5]</a></sup>'],'',regex=True)
web_parse_after
```

# Dealing with Outliers

```python
sale_join3.describe()
```

```python
mean = sale_join3.describe().iloc[1,:]
mean
```

```python
first_Q = sale_join3.describe().iloc[4,:]
third_Q = sale_join3.describe().iloc[6,:]
print(first_Q)
print(third_Q)
```

```python
IQR = third_Q - first_Q
IQR
```

```python
upper = mean + 1.5*IQR
lower = mean - 1.5*IQR
print(upper)
print(lower)
```

### Outlier conclusion
From the data, the outliers in each category are shown below:

Sale: Cheyenne, Gillette ----- Too high
      Buffalo, Douglas ------- Too low
Land area: Rock Springs ------ Too high
Households with under 18: Casper ------ Too high
Population density: Cheyenne -------Too high
Total Families: Cheyenne ---------Too high
2010 Census: Cheyenne -------Too high

From the results above, because Cheyenne is an outlier in 4 different categories, Cheyenne will be deleted from the analysis since it's an outlier.
